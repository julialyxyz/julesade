### FizzBuzz Function
 
**Create a function fizzbuzz(n) that takes an integer n as a parameter.**

The function prints: 
* "Fizz" for multiples of 3 
* "Buzz" for multiples of 5 
* "FizzBuzz" for multiples of both 3 and 5
* The number itself for other numbers

```python
def fizzbuzz(n):         
    x = n%3
    y = n%5
    if x == 0 and y == 0:
        return "FizzBuzz"
    elif  x == 0:
        return "Fizz"
    elif y == 0:
        return "Buzz"
    else:
        return n
```

**Call FizzBuzz function for numbers 1 to 20.**

```python
for i in range(1,21):
    print(fizzbuzz(i))
```
