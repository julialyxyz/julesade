```python
#import titanic dataset
path = '/Users/julia/Desktop/ESADE/python/assignment_6/train_and_test2.csv'
df = pd.read_csv(path)
```

<div class="alert alert-block alert-info">
<b>exercise 1:</b> Calculate the gender-based survival percentage
</div>


```python
#calculate survival percentage for male
male = df[(df['Sex'] == 0)]['2urvived'].value_counts(normalize=True)
male
```

```python
#calculate survival percentage for female
female = df[(df['Sex'] == 1)]['2urvived'].value_counts(normalize=True)
female
```

```python
print(f"survival percentage: \nmale: {round(male.iloc[1]*100,1)}% \nfemale: {round(female.iloc[1]*100,1)}%")
```

<div class="alert alert-block alert-info">
<b>exercise 2:</b> Calculate the survival percentage grouped by gender and class
</div>


```python
#group by passenger class and sex, then calculate survival rates
survival = pd.DataFrame(df.groupby(["Pclass","Sex"])["2urvived"].value_counts(normalize=True))
survival
```

```python
#reset the index
survival = survival.reset_index()

#filter for rows where passenger survived as proportions
survival = survival[survival["2urvived"] == 1]

#convert the survival proportions to percentage
survival["proportion (%)"] = round(survival["proportion"]*100,2)

#drop the survival and proportion columns
survival.drop(["2urvived", "proportion"], axis=1, inplace = True)
survival
```
